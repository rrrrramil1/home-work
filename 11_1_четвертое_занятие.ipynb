```python
def AddRightDigit(D, K):
    if 0 <= D <= 9:
        return K * 10 + D
    else:
        print("Ошибка: D должно быть цифрой (0-9)")
        return K

K = int(input("Введите число K: "))
D1 = int(input("Введите цифру D1 (0-9): "))
D2 = int(input("Введите цифру D2 (0-9): "))

K = AddRightDigit(D1, K)
print("После добавления D1:", K)
K = AddRightDigit(D2, K)
print("После добавления D2:", K)
```

```python
def AddLeftDigit(D, K):
    if 1 <= D <= 9:
        return D * (10 ** len(str(K))) + K
    else:
        print("Ошибка: D должно быть цифрой (1-9)")
        return K

K = int(input("Введите число K: "))
D1 = int(input("Введите цифру D1 (1-9): "))
D2 = int(input("Введите цифру D2 (1-9): "))

K = AddLeftDigit(D1, K)
print("После добавления D1:", K)
K = AddLeftDigit(D2, K)
print("После добавления D2:", K)
```

```python
def Swap(X, Y):
    return Y, X

A = float(input("Введите A: "))
B = float(input("Введите B: "))
C = float(input("Введите C: "))
D = float(input("Введите D: "))

A, B = Swap(A, B)
C, D = Swap(C, D)
B, C = Swap(B, C)

print("Новые значения:")
print("A =", A)
print("B =", B)
print("C =", C)
print("D =", D)
```

```python
def Minmax(X, Y):
    if X > Y:
        return Y, X
    else:
        return X, Y

A = float(input("Введите A: "))
B = float(input("Введите B: "))
C = float(input("Введите C: "))
D = float(input("Введите D: "))

A, B = Minmax(A, B)
C, D = Minmax(C, D)
A, C = Minmax(A, C)
B, D = Minmax(B, D)

print("Минимальное:", A)
print("Максимальное:", D)
```

```python
def SortInc3(A, B, C):
    # Сортировка пузырьком для трех элементов
    if A > B: A, B = B, A
    if B > C: B, C = C, B
    if A > B: A, B = B, A
    return A, B, C

# Первый набор
A1 = float(input("Введите A1: "))
B1 = float(input("Введите B1: "))
C1 = float(input("Введите C1: "))
A1, B1, C1 = SortInc3(A1, B1, C1)
print("Первый набор после сортировки:", A1, B1, C1)

# Второй набор
A2 = float(input("Введите A2: "))
B2 = float(input("Введите B2: "))
C2 = float(input("Введите C2: "))
A2, B2, C2 = SortInc3(A2, B2, C2)
print("Второй набор после сортировки:", A2, B2, C2)
```
