```python
N = int(input("Введите N (>2): "))
if N <= 2:
    print("Ошибка: N должно быть >2")
else:
    fib = [1, 1]
    for i in range(2, N):
        fib.append(fib[i-2] + fib[i-1])
    print("Последовательность Фибоначчи:", fib)
```

```python
def matrix11():
    # Пример матрицы 4x4
    matrix = [
        [11, 12, 13, 14],
        [21, 22, 23, 24],
        [31, 32, 33, 34],
        [41, 42, 43, 44]
    ]

    print("Исходная матрица:")
    for row in matrix:
        print(row)

    print("\nЗмейка по строкам:")
    for i in range(len(matrix)):
        if i % 2 == 0:
            print(*matrix[i])  # Четные строки слева направо
        else:
            print(*matrix[i][::-1])  # Нечетные строки справа налево

# Запуск задачи
matrix11()
```

```python
def matrix12():
    # Пример матрицы 3x5
    matrix = [
        [11, 12, 13, 14, 15],
        [21, 22, 23, 24, 25],
        [31, 32, 33, 34, 35]
    ]

    print("Исходная матрица:")
    for row in matrix:
        print(row)

    print("\nЗмейка по столбцам:")
    for j in range(len(matrix[0])):
        if j % 2 == 0:
            # Четные столбцы сверху вниз
            column = [matrix[i][j] for i in range(len(matrix))]
        else:
            # Нечетные столбцы снизу вверх
            column = [matrix[i][j] for i in range(len(matrix)-1, -1, -1)]
        print(f"Столбец {j+1}:", column)

# Запуск задачи
matrix12()
```

```python
def matrix13():
    # Квадратная матрица 4x4
    matrix = [
        [11, 12, 13, 14],
        [21, 22, 23, 24],
        [31, 32, 33, 34],
        [41, 42, 43, 44]
    ]

    print("Исходная матрица:")
    for row in matrix:
        print(row)

    print("\nВывод уголками (вариант 1):")

    m = len(matrix)
    result = []
    top, bottom = 0, m-1
    left, right = 0, m-1

    while top <= bottom and left <= right:
        # Верхняя строка слева направо
        for j in range(left, right+1):
            result.append(matrix[top][j])
        top += 1

        # Правый столбец сверху вниз (без первого)
        for i in range(top, bottom+1):
            result.append(matrix[i][right])
        right -= 1

        if top <= bottom:
            # Нижняя строка справа налево
            for j in range(right, left-1, -1):
                result.append(matrix[bottom][j])
            bottom -= 1

        if left <= right:
            # Левый столбец снизу вверх
            for i in range(bottom, top-1, -1):
                result.append(matrix[i][left])
            left += 1

    print("Результат:", ' '.join(map(str, result)))

# Запуск задачи
matrix13()
```

```python
def matrix14():
    # Квадратная матрица 4x4
    matrix = [
        [11, 12, 13, 14],
        [21, 22, 23, 24],
        [31, 32, 33, 34],
        [41, 42, 43, 44]
    ]

    print("Исходная матрица:")
    for row in matrix:
        print(row)

    print("\nВывод уголками (вариант 2):")

    m = len(matrix)
    result = []
    top, bottom = 0, m-1
    left, right = 0, m-1

    while top <= bottom and left <= right:
        # Левый столбец сверху вниз
        for i in range(top, bottom+1):
            result.append(matrix[i][left])
        left += 1

        # Нижняя строка слева направо (без первого)
        for j in range(left, right+1):
            result.append(matrix[bottom][j])
        bottom -= 1

        if left <= right:
            # Правый столбец снизу вверх
            for i in range(bottom, top-1, -1):
                result.append(matrix[i][right])
            right -= 1

        if top <= bottom:
            # Верхняя строка справа налево
            for j in range(right, left-1, -1):
                result.append(matrix[top][j])
            top += 1

    print("Результат:", ' '.join(map(str, result)))

# Запуск задачи
matrix14()
```
