```python
def is_lucky(ticket):
    digits = [int(d) for d in str(ticket).zfill(6)]
    return sum(digits[:3]) == sum(digits[3:])

N = int(input("Введите начальный номер билета (6 цифр): "))
M = int(input("Введите конечный номер билета (6 цифр): "))

count = 0
for ticket in range(N, M + 1):
    if is_lucky(ticket):
        count += 1

print(f"Количество счастливых билетов: {count}")
```

```python
import math

N = int(input("Введите размер массива N: "))
X = [1 / math.factorial(n) for n in range(1, N + 1)]

print("Полученный массив:")
for i, x in enumerate(X, 1):
    print(f"X[{i}] = {x:.6f}")
```

```python
initial = 1  # Начальное количество амеб
hours = 24    # Общее время наблюдения

print("Количество амеб:")
for time in range(3, hours + 1, 3):
    cells = initial * (2 ** (time // 3))
    print(f"Через {time} часов: {cells} клеток")
```
